```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Aman", "Riya", "Karan", "Neha", "Rahul"],
    "Marks": [85, 90, 78, 92, 88],
    "Subject": ["Math", "Science", "Math", "Science", "Math"]
}

df = pd.DataFrame(data)
```

```python
data
```

```python
df.head(3)
```

```python
df.describe()
```

```python
avg=df["Marks"].mean()
print(avg)
```

```python
# Score greater than 85
great=df[df["Marks"]>85]
print(great)
```

```python
print(df["Subject"].value_counts())
```

```python
df["Result"] = ["Pass" if x >= 80 else "Fail" for x in df["Marks"]]
```

```python
df
```

```python
df_sorted = df.sort_values(by="Marks", ascending=False)
print(df_sorted)
```
