**Some Important Functions in NumPy**
- `np.max(x)` - Returns the maximum value in an array.
- `np.min(x)` - Returns the minimum value in an array.
- `np.argmax(x)` - Returns the index of the maximum value in an array.
- `np.argmin(x)` - Returns the index of the minimum value in an array.
